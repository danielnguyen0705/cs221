```python
import nltk
from nltk.probability import LaplaceProbDist, WittenBellProbDist
from nltk.probability import LidstoneProbDist, SimpleGoodTuringProbDist
from nltk.tag.hmm import HiddenMarkovModelTrainer
from nltk.tag import UnigramTagger


nltk.download("treebank")
nltk.download("punkt_tab")
```

Ngữ liệu gán nhãn từ loại. Một phần của Penn Treebank

```python
data = nltk.corpus.treebank.tagged_sents()
train = data[:3000]
test = data[3000:]
```

Most Frequent Class

```python
Utagger = UnigramTagger(train)
print("Unigram Tagger accuracy:", Utagger.accuracy(test))
```

Hidden Markov không sử dụng smoothing

```python
trainer = HiddenMarkovModelTrainer()
tagger = trainer.train_supervised(train)
```

```python
tagger.accuracy(test)
```

Hidden Markov sử dụng Laplace smoothing

```python
HMMtrainer = HiddenMarkovModelTrainer()
HMMtagger = HMMtrainer.train_supervised(train, estimator=LaplaceProbDist)

print("HMM accuracy:", HMMtagger.accuracy(test))
```

Hidden Markov sử dụng Witten-Bell smoothing

```python
HMMtrainer = HiddenMarkovModelTrainer()
HMMtagger = HMMtrainer.train_supervised(train, estimator=WittenBellProbDist)

print("HMM accuracy:", HMMtagger.accuracy(test))
```
